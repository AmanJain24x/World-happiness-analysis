## PREDICTIVE ANALYSIS

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
Report_2021.head()
```

```python
LM_2021=Report_2021.copy()
```

```python
x1=LM_2021.drop(['Country name','Regional indicator','Rank','Ladder score'],axis=1,inplace=False)
```

```python
y1=LM_2021['Ladder score']
```

#To check normality:

```python
LM_2021['Ladder score'].hist()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.3,random_state=3)
```

```python
LinearModel=LinearRegression(fit_intercept=True)
```

```python
Model_LinR=LinearModel.fit(x_train,y_train)
```

```python
pred_y=LinearModel.predict(x_test)
```

```python
pred_y
```

```python
LinearModel.coef_
```

```python
LinearModel.intercept_
```

```python
r2_test=Model_LinR.score(x_test,y_test)
```

```python
print(r2_test)
```

```python
r2_train=Model_LinR.score(x_train,y_train)
```
